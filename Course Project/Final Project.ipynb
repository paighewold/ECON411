### Total

```python
def total(lst):
    total_ = 0
    for val in lst:
        total_ += val
    return total_
```

### Mean

```python
def mean(lst):
    n = len(lst)
    mean_ = total(lst)/n
    return mean_
```

### Median

```python
def median(lst):
    n = len(lst)
    lst = sorted(lst)
    
    # two cases:
    #   1. list of odd length
    #   i % j checks for remainder upon dividing i by j
    if n % 2 != 0:
        middle_index = int((n-1)/2)
        median_ = lst[middle_index]
        
    else:
        upper_middle_index = int(n/2)
        lower_middle_index = upper_middle_index - 1
        # pass slice with two middle values to mean()
        median_ = mean(lst[lower_middle_index: upper_middle_index + 1])
        
    return median_
```

### Mode

```python
def mode(lst):
    count_dct = {}
    # create entries for each value with 0
    for key in lst:
        count_dct[key] = 0
    # add up each occurence
    for key in lst:
        count_dct[key] += 1
    
    #calculate max_count up front
    max_count = max(count_dct.values())
    #now we can compare each count to the max count
    mode_ = []
    for key, count in count_dct.items():
        if count == max_count:
            mode_.append(key)
    
    return mode_
```

### Variance

```python
def variance(lst, sample =True):
    list_mean = mean(lst)
    n = len(lst)
    DoF = n - 1
    sum_sq_diff = 0
    
    for val in lst:
        diff = val - list_mean
        sum_sq_diff += (diff)**2
 #       print(val, list_mean, diff**2, sum_sq_diff)
    if sample == False:
        variance_ = sum_sq_diff/n
    else:
        variance_ = sum_sq_diff/DoF
    return variance_
```

### Standard Deviation

```python
def SD(lst, sample = True):
    SD_ = variance(lst, sample) ** (1/2)
    return SD_
```

### Standard Error

```python
def STE(lst, sample = True):
    n = len(lst)
    se = SD(lst, sample) / n ** (1/2)
    
    return se
```

### Covariance

```python
def covariance(lst1, lst2, sample = True):
    mean1 = mean(lst1)
    mean2 = mean(lst2)
    
    cov = 0
    n1 = len(lst1)
    n2 = len(lst2)
    if n1 == n2:
        n = n1
        # sum the product of the differences
        for i in range(n):
            cov += (lst1[i] - mean1) * (lst2[i] - mean2)
        if sample == False:
            cov = cov/n
        else:
            cov = cov/(n-1)
            
        return cov
    else:
        print("List lengths are not equal")
        print("List1: ", n1)
        print("List2: ", n2)
```

### Correlation

```python
def correlation(lst1, lst2):
    cov = covariance(lst1, lst2)
    SD1 = SD(lst1)
    SD2 = SD(lst2)
    corr = cov / (SD1 * SD2)
    
    return corr
```

### Skewness

```python
def skewness(lst, sample = True):
    mean_ = mean(lst)
    SD_ = SD(lst, sample)
    skew = 0
    n = len(lst)
    for val in lst:
        skew += (val - mean_) ** 3
    skew = skew/(n*SD_**3) if not sample else\
    n * skew / ((n-1)*(n-2)*SD_**3)
    
    return skew
```

### Kurtosis

```python
def kurtosis(lst, sample = True):
    mean_ = mean(lst)
    kurt = 0
    SD_ = SD(lst, sample)
    n = len(lst)
    for val in lst:
        kurt += (val - mean_) ** 4
    kurt = kurt / (n * SD_ ** 4) if  sample == False else  n * (n + 1) * kurt / \
    ((n - 1) * (n - 2) * (n - 3) * (SD_ ** 4)) - (3 *(n - 1) ** 2) / ((n - 2) * (n - 3))
    
    return kurt
```

```python
def gather_statistics(df, sample = False, round_dig = 3):
    dct = {key:{} for key in df}
    for key, val in df.items():
        val.dropna(inplace = True)
        dct[key]["mean"] = round(mean(val), round_dig)
        dct[key]["median"] = round(median(val), round_dig)
        dct[key]["variance"] = round(variance(val), round_dig)
        dct[key]["S.D."] = round(SD(val, sample), round_dig)
        dct[key]["Skewness"] = round(skewness(val, sample), round_dig)
        dct[key]["Kurtosis"] = round(kurtosis(val, sample), round_dig)

    stats_df = pd.DataFrame(dct)
    return stats_df
```

## Determining and Removing Outliers

Outliers are not good in a statistical analysis because they can have a significant impact on summary statistics and can distort the interpretation of statistical results. Outliers can arise for many reasons, such as measurement error, data entry errors, or real differences in the sample size.
