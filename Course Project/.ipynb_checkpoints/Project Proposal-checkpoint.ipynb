# Project Proposal

In recent years, bike sharing has emerged as a popular mode of transportation that offers many benefites, including convenience, sustainability, and cost savings. Bike sharing systems allow users to rent bicycles from designated stations and return them to any other station within the network. The popularity of bike sharing systems has led to the rapid expansion of these systems across the world, making them a vital component of modern urban transportation systems.

This study aims to analyze the factors affecting bike sharing demand using the bike sharing dataset collected from the Capital Bikeshare system. The dataset consists of bike rental information for the years 2011 and 2012, including the date and time of rental weather conditions, and various other demographic and situational factors. The purpose of this study is to understand the relationship between these factors nd bike rental demand and to provide valuable insights into the factors affecting bike sharing demand.

Regression analysis is a statistical method used to explore the relationship between a dependent variable and one or more independent variables. In the context of the bike sharing dataset, regression analysis can be used to explore the relationship between bike rental demand and various factors, such as weather conditions, seasonal patterns, time of day, day of the week, and location of bike rental stations. By performing regression analysis on the bike sharing dataset, it becommes possible to quantify the strength and direction of the relationship between these factors and bike rental demand.

This can provide valuable insights into the factors affecting bike sharing demand and can help to make the data easier to read by summarizing the relationship between multiple variables into a single model. The regression model can be used to predict bike rental demand based on the values of the independent variables, which can help to identify the factors that are most important in determining bike rental demand. This information can then be used to inform future planning and policy decisions related to bike sharing systems.

Futhermore, the regression analysis can also provide information on the statistical significance of each independent variable, which can help to determine which factors are actually having an impact on bike rental demand and which factors are not. This information can help to make the data easier to interpret and to understand the most importants drivers of bike sharing demand.

One economic theory that can be applied to this dataset is the theory of demand and supply. According to this theory, the demand for bike sharing services will increase as the price of the service decreases and the quality of the service increases. This theory can be applied to the bike sharing dataset by analyzing the relationship between bike rental prices, the quality of the service, and bike rental demand. For examples, if the price of bike rental services decreases, demand for bike sharing services is expected to increase. This relationship can be explored using regression analysis, which I talked about earlier.

Another economic theory that can be applied to this dataset is the theory of consumer behavior. According to this theory, consumer behavior is influenced by various factors, including income, preferences, and availability of alternatives. This theory can be applied to the bike sharing dataset by analyzing the relationship between bike rental demand and factors such as weather conditions, time of day, day of the week, and location of bike rental stations. For example, if the weather is sunny and warm, demand for bike sharing services is expected to increase, as consumers may be more likely to choose bike sharing as a mode of transportation. This relationship can also be explored using regression analysis.

Bike sharing can also have positive effects on the economy. Bike sharing programs can create jobs in a variety of fields, including bike maintenance, management, and customer service. It can also reduce transportation costs. By providing a low-cost alternative to traditional modes of transportation, bike sharing can help to reduce transportation costs for individuals and households. This can free up disposable income for other expenditures, which can boost consumer spending and support economic growth.

On the other hand, bike sharing can have a negative effect on the economy. Bike sharing programs can compete with traditional transportation providers, such as taxis, buses, and trains. This competition can lead to reduced profits and job losses in these industries. Also, the costs of implementing and maintaining a bike sharing program can be substantial, which can reduce the resources available for other economic activities.

Overall, the bike sharing dataset can be analyzed using various economic theories to gain a better understanding of the factors affecting bike sharing demand. The theories of demand and supply and consumer behavior can be applied to the dataset by analyzing the relationship between bike rental demand and various factors, including price, quality of the service, weather conditions, time of day, day of the week, and location of bike rental stations.

```python
import pandas as pd
df = pd.read_csv("day.csv")
df
```

```python
bikeshare_df = pd.DataFrame(df)
bikeshare_df
```

```python
import matplotlib.pyplot as plt
bikeshare_df.plot()
```

```python
fig, ax = plt.subplots(figsize = (20,12))
bikeshare_df.plot(linewidth = 5, alpha = 0.7, ax=ax) # key words
ax.legend(fontsize = 20)
```

```python
bikeshare_df = bikeshare_df.rename(columns = {"yr": "Year", "mnth": "Month", "dteday": "Date", 
                                              "holiday": "Holiday", "cnt": "# of Bikes Rented"})
bikeshare_df
```

```python
plt.rcParams.update({"font.size": 20})
plot_df = bikeshare_df[["Date", "# of Bikes Rented"]].iloc[1:]
fig, ax = plt.subplots(figsize = (40,20))
plot_df.plot.bar(secondary_y = "Date",
                width = 0.9,
                alpha = 0.75,
                color = ["C3", "C1"],
                ax = ax)
x_ticklabels = [str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
ax.set_title("Daily Count of Bikes Rented")
```
